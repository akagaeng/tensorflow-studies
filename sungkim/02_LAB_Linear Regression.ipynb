$$H(x) = W_x + b$$


$$cost(W) = \frac{1}{m}\sum_{i=1}^{m} (W_x ^{(i)} - y^{(i)})^2$$


- y: true value
- H(x): 예측, 가설(Hypothesis)
- 잘 예측할수록 cost function 작아짐
- 학습을 한다는 것은 W와 b를 조정하여 cost function cost(W,b)을 최소화하는 것!



### 1. Build graph using TF operations

```python
import tensorflow as tf

# H(x) = Wx + b

# X, Y data (우리가 사용하는 데이터 형태)
x_train = [1,2,3]
y_train = [1,2,3]

# tf.Variable
# - 텐서플로가 사용하는 데이터형태, 일반적인 변수와는 다름
# - trainable variable 이라고도 함

# 텐서플로 변수 선언
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = x_train * W + b

# cost(loss) function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))

# Gradient Descent
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost) # optimizer를 통해 cost를 minimize하라!

# 그래프 구현 완료
```

### 2. Run/Update graph and get results

```python
# 그래프를 실행하기 위해 session을 만들어야 함
# tensorflow variable (W, b, ...)을 사용하기 위해서는 초기화해주어야 함)
# by tf.global_variables_initializer

# Launch the graph in a session
sess = tf.Session()

# Initializes global variables in the graph
sess.run(tf.global_variables_initializer())

# Fit the line
for step in range(2001):
    sess.run(train) # 그래프의 train node를 실행시킴
    if step % 10 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

# 결과
- 실제 정답은 W=1, b=0, cost = 0
- step이 증가할수록 cost값이 작아지며, W, b가 특정 값(1)으로 수렴함(소수점 확인 잘 할 것)
- train node를 실행시킨다는 것은 train - cost - hypothesis - W, b 로 연결된 그래프를 실행시켜 W, b까지 값을 조정할 수 있음

### Placeholder 사용하여 cost function 구하기

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = X * W + b

# cost(loss) function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Gradient Descent
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost) # optimizer를 통해 cost를 minimize하라!

# Launch the graph in a session
sess = tf.Session()

# Initializes global variables in the graph
sess.run(tf.global_variables_initializer())

# Fit the line
for step in range(2001):
    # train된 값은 _ 변수에 담음: 버리는 변수 담을 때 관용적으로 사용하는 듯
    cost_val, W_val, b_val, _ = \
    sess.run([cost, W, b, train], feed_dict = \
             {X: [1,2,3], Y: [1,2,3]}) 
    # X, Y 값이 미리 정해져있지 않더라도 run시킬 때 지정하여 실행시킬 수 있음!
    if step % 10 == 0:
        print(step, cost_val, W_val, b_val)
```

# Placeholder 사용하는 이유
